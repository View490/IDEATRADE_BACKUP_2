```python
# READ DF1 THEN PUSH DATA INTO TABLE Program_Trading_SET_and_Mai

import pandas as pd
import mysql.connector
from datetime import datetime

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("df1.csv")
print('df len: ', df.__len__)

# Function to convert datetime strings to MySQL-compatible format
def convert_to_mysql_datetime(datetime_str):
    try:
        dt = datetime.strptime(datetime_str, "%A, %B %d, %Y")
        return dt.strftime("%Y-%m-%d %H:%M:%S")
    except ValueError:
        return None  # Return None if datetime string cannot be parsed

# Convert datetime column to MySQL-compatible format
df['datetime'] = df['datetime'].apply(convert_to_mysql_datetime)

# Establish connection to MySQL server
conn = mysql.connector.connect(
    host="ideatrade.serveftp.net",
    user="Sura@view",
    password="ZP6HpF-T04CGw_t(",
    database="db_ideatrade",
    port=51410
)

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# Define column names and their data types
column_definitions = [
    "`datetime` DATETIME",
    "`Program_Trading_Buy` FLOAT",
    "`Program_Trading_Sell` FLOAT",
    "`Program_Trading_Net` FLOAT",
    "`Non_Program_Trading_Buy` FLOAT",
    "`Non_Program_Trading_Sell` FLOAT",
    "`Non_Program_Trading_Net` FLOAT",
    "`Total_Buy` FLOAT",
    "`Total_Sell` FLOAT"
]

# Define your SQL query to create a new table
table_name = "Program_Trading_SET_and_Mai"
create_table_query = """
CREATE TABLE IF NOT EXISTS {} (
    {}
)
""".format(table_name, ',\n    '.join(column_definitions))

# Execute the SQL query
cursor.execute(create_table_query)

# Iterate through DataFrame rows and insert data into MySQL table
for _, row in df.iterrows():
    insert_query = """
    INSERT INTO {} ({}) VALUES ({})
    """.format(table_name, ','.join(['`' + col.replace(' ', '_') + '`' for col in df.columns]), ','.join(['%s']*len(df.columns)))
    cursor.execute(insert_query, tuple(row))

# Commit changes
conn.commit()

# Close cursor and connection
cursor.close()
conn.close()
```

```python
# Pulling Symbol Closing Price Data
# SELECT PROGRAM_TRADING_BY_SECURITIES TABLE >>> PRINT COLUMNS NAME

import mysql.connector
name = 'stock_hloc'
# SEELCT PROGRAM TRADING BY SECURITIES TABLE >>> PRINT VALUES

conn = mysql.connector.connect(
    host = 'ideatrade.serveftp.net',
    user = 'Sura@view',
    password = 'ZP6HpF-T04CGw_t(',
    database = 'db_ideatrade',
    port = 51410
    )

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# Define the table name with spaces
table_name = 'stock_hloc'

##################################### FILTER TRAINFORMS AND SAVE
import json

# Define your SQL query to select distinct stock names
select_stock_query = 'SELECT DISTINCT StockList FROM `{}`'.format(table_name)
print('SQL to fetch stock names:', select_stock_query)

# Execute the query to fetch distinct stock names
cursor.execute(select_stock_query)

# Fetch all distinct stock names
stock_names = cursor.fetchall()
# stock_names = stock_names[:100]
print('Stock Names len: ', len(stock_names))
# stock_names = [("BYD",),("ZIGA",),("BAM",),("SAT",)]

# Initialize closing_prices dictionary with labels and stocks
closing_prices = {'labels': []}


from tqdm import tqdm
# Iterate over each stock name to fetch data
for idx, stock_name in tqdm(enumerate(stock_names)):
    # Extract the stock name from the fetched data
    target_stock = stock_name[0]

    # Define your SQL query to select rows for the target stock
    select_query = 'SELECT Time, Close FROM `{}` WHERE StockList = "{}" AND TimeFrame = "1D" ORDER BY date DESC LIMIT 150'.format(table_name, target_stock)
    # print('SQL to fetch data for', target_stock, ':', select_query)

    # Execute the query to fetch data for the target stock
    cursor.execute(select_query)

    # Fetch all rows for the target stock
    rows = cursor.fetchall()

    # Initialize lists for storing time and close values
    time_values = []
    close_values = []

    # Iterate over each row in the fetched data
    for row in rows:
        # Extract values from the row
        time = row[0].strftime("%A, %B %d, %Y")  # Format datetime as "Wednesday, March 6, 2024"
        print('row[0]: ',row[0])
        close = row[1]

        # Append time and close values to the respective lists
        time_values.append(time)
        close_values.append(close)

    # Append time and close values to the closing_prices dictionary
    closing_prices['labels'] = time_values  # Labels are common for all stocks
    closing_prices[target_stock] = close_values 
    # Write the converted JavaScript data to a file
    
with open('closing_prices.js', 'w') as file:
    file.write('var closing_prices = ' + json.dumps(closing_prices, indent=4) + ';')
    

# Close cursor and connection
cursor.close()
conn.close()


```
